```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp

fontsize = 18
```

```python
# Exploring Exponential Signals

# Signal Plot of Aα^n when A is +ve and α is between 0 and 1
a = 10 # Amplitude
alpha = 0.65
n = np.array([x for x in np.arange(0, 15, 1)])
y = a * alpha ** n

# Plot figure
fig, ax = plt.subplots(figsize = (13, 5))
ax.set_xlabel('n'), ax.set_xlim(-1, 15)
ax.set_ylabel('Amplitude')
ax.set_title('Signal Plot: $Aα^{n}$ when  A is +ve and α is between 0 and 1')
ax.stem(n, y)
```

```python
# Signal Plot of Aα^n when A is +ve and α is between -1 and 0
a = 10 # Amplitude
alpha2 = -0.65 # Negative exponential
n = np.array([x for x in np.arange(0, 15, 1)])
y = a * alpha2 ** n

# Plot figure
fig2, ax2 = plt.subplots(figsize = (13, 5))
ax2.set_xlabel('n'), ax.set_xlim(-1, 15)
ax2.set_ylabel('Amplitude')
ax2.set_title('Signal Plot: $Aα^{n}$ when  A is +ve and α is between -1 and 0')
ax2.stem(n, y)
```

```python
# Investigating Periodicity for discrete-time signals

n = np.array([x for x in np.arange(0, 41, 1)])
period = 8

# Signal equations
x1_n = np.cos(np.pi * n / 4)
x1_n_N = np.cos(np.pi * (n + period) / 4)

x2_n = np.cos((3 * np.pi * n) / 8)
x2_n_N = np.cos((3 * np.pi * (n + period)) / 8)

# Plot figure
fig2, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (20, 15))

# plot x1_n and x2_n
ax1.stem(n, x1_n)
ax1.plot(n, x1_n, 'r--') # Trace the plots
ax1.set_title('x1[n] = cos(πn/4)', fontsize=fontsize)
ax2.stem(n, x2_n)
ax2.plot(n, x2_n, 'r--')
ax2.set_title('x2[n] = cos(3πn/8)', fontsize=fontsize)

# plot x1_n_N and x2_n_N
ax3.stem(n, x1_n_N)
ax3.plot(n, x1_n, 'r--') # Trace x1_n over x1_n_N
ax3.set_title('x1[n+8] = cos(π(n+8)/4)', fontsize=fontsize)
ax4.stem(n, x2_n_N)
ax4.plot(n, x2_n, 'r--') # Trace x2_n over x2_n_N
ax4.set_title('x2[n+8] = cos(3π(n+8)/8)', fontsize=fontsize)
```

Diagram above shows that 8 is a period for $x1[n] = cos(\frac{πn}{4})$ but not for $x2[n] = cos(\frac{3πn}{8})$
